```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
from IPython.display import Image
from skimage.util import random_noise
```

```python
image = cv2.imread('/content/Saul.jpeg')
```

**Original Image**

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Original image')
plt.imshow(img)
plt.axis('off')
plt.show()
```

**Median Salt & Pepper denoise**
> * cv2.medianBlur takes the image and the kernal size and applies medianBlur<br>
> * the function return the blurred image

```python
def Median_Blur(img,kernal):
      S_P= cv2.medianBlur(img, kernal)
      return S_P
blur=Median_Blur(image,3)
blur=cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Median Blur')
plt.imshow(blur)
plt.axis('off')
plt.show()
```

**Pepper(Max) denoise**
>* iniate a 2d kernel of size of kernal kernal using np.ones function to
make a 2d array of ones
>* np.unit8 declares the data type of the kernel is unsigned 8bit int
>* dilate is a builtin function for Max blurring which removes the pepper noise
>* the function return the blurred image

```python
def Mx(img,kernal):
      kernel = np.ones((kernal,kernal),np.uint8)
      pepper = cv2.dilate(img, kernel)
      return pepper
blur=Mx(image,3)
blur=cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Max blur')
plt.imshow(blur)
plt.axis('off')
plt.show()
```

**Salt (Min) denoise**
> * iniate a 2d kernel of size of kernal * kernal using np.ones function to
make a 2d array of ones <br>
> * np.unit8 declares the data type of the kernel is unsigned 8bit int<br>
> * erode is a builtin function for Min blurring which removes the salt noise<br>
> * the function return the blurred image

```python
def Mn(img,kernal):
      kernel = np.ones((kernal,kernal),np.uint8)
      salt = cv2.erode(img, kernel)
      return salt
blur=Mn(image,3)
blur=cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Min Blur')
plt.imshow(blur)
plt.axis('off')
plt.show()
```

**Mean**
> * cv2.blur takes the image and the kernal size and applies mean blur <br>
> * the function return the blurred image

```python
def mean(img,kernel):
  mean_filtered = cv2.blur(img, (kernel, kernel))
  return mean_filtered
blur=mean(image,3)
blur=cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Mean blur')
plt.imshow(blur)
plt.axis('off')
plt.show()
```

**Gaussian Blur**
>* cv2.GaussianBlur takes img and kernel size and sigmaX = 0 and sigmaY = None <br>
> * where sigmaX and sigmaY are standard deviation in direction x and y respectively<br>

```python
def Gaussian(img,kernel):
  gauss = cv2.GaussianBlur(img, (kernel,kernel), 0)
  return gauss
blur=Gaussian(image,3)
blur=cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Gaussian blur')
plt.imshow(blur)
plt.axis('off')
plt.show()
```
