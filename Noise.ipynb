```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
from IPython.display import Image
from skimage.util import random_noise
```

```python
image = cv2.imread('/content/Saul.jpeg')
```

**Original image**

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Original image')
plt.imshow(img)
plt.axis('off')
plt.show()
```

**Add Salt and Pepper noise**
>* noisy_image = image.copy() makes a copy of the original image so the input is not changed.
>*  salt_mask = np.random.random(image.shape) < salt_prob creates a mask where random pixels are chosen to become salt (white).
>* salt_prob controls how many pixels are affected.
>* noisy_image[salt_mask] = 255 sets the selected pixels in the mask to 255 (white).
>* pepper_mask = np.random.random(image.shape) < pepper_prob creates another mask for pepper (black) pixels.
>* pepper_prob controls how many pixels are affected.
>* noisy_image[pepper_mask] = 0 sets the selected pixels in the pepper mask to 0 (black).
>* return noisy_image returns the new image with salt-and-pepper noise applied.

```python
def add_salt_pepper_noise(image, salt_prob=0.01, pepper_prob=0.01):
    noisy_image = image.copy()
    salt_mask = np.random.random(image.shape) < salt_prob
    noisy_image[salt_mask] = 255
    pepper_mask = np.random.random(image.shape) < pepper_prob
    noisy_image[pepper_mask] = 0
    return noisy_image
noise=add_salt_pepper_noise(image,0.15,0.15)
noise=cv2.cvtColor(noise,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('S&P Noise')
plt.imshow(noise)
plt.axis('off')
plt.show()
```

**Add Salt noise**
>* noisy_image = image.copy() makes a copy of the original image so the input is not changed.
>*  salt_mask = np.random.random(image.shape) < salt_prob creates a mask where random pixels are chosen to become salt (white).
>* salt_prob controls how many pixels are affected.
>* noisy_image[salt_mask] = 255 sets the selected pixels in the mask to 255 (white).
>* return noisy_image returns the new image with salt noise applied.

```python
def add_salt_noise(image, salt_prob=0.01):
    noisy_image = image.copy()
    salt_mask = np.random.random(image.shape) < salt_prob
    noisy_image[salt_mask] = 255
    return noisy_image
noise=add_salt_noise(image,0.15)
noise=cv2.cvtColor(noise,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Salt Noise')
plt.imshow(noise)
plt.axis('off')
plt.show()
```

**Add Pepper noise**
>* noisy_image = image.copy() makes a copy of the original image so the input is not changed.
>* pepper_mask = np.random.random(image.shape) < pepper_prob creates another mask for pepper (black) pixels.
>* pepper_prob controls how many pixels are affected.
>* noisy_image[pepper_mask] = 0 sets the selected pixels in the pepper mask to 0 (black).
>* return noisy_image returns the new image with pepper noise applied.

```python
def add_pepper_noise(image, pepper_prob=0.01):
    noisy_image = image.copy()
    pepper_mask = np.random.random(image.shape) < pepper_prob
    noisy_image[pepper_mask] = 0
    return noisy_image
noise=add_pepper_noise(image,0.15)
noise=cv2.cvtColor(noise,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.title('Pepper Noise')
plt.imshow(noise)
plt.axis('off')
plt.show()
```
